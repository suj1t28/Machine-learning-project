```python
#Outliers
```

```python
per25=df['degree_p'].quantile(0.25)
```

```python
per75=df['degree_p'].quantile(0.75)
```

```python
iqr=per75-per25
```

```python
iqr
```

```python
upperlimit=per75+(1.5*iqr)
lowerlimit=per25-(1.5*iqr)
```

```python
upperlimit
```

```python
lowerlimit
```

```python
df=df[(df['degree_p']>lowerlimit) & (df['degree_p']<upperlimit)]
```

```python
df
```

```python
sns.boxplot(y='degree_p',data=df)
```

```python
df.columns
```

```python
col=['ssc_p','hsc_p','etest_p','mba_p','salary']
for c in col:
    per25=df[c].quantile(0.25)
    per75=df[c].quantile(0.75)
    iqr=per75-per25
    upperlimit=per75+(1.5*iqr)
    lowerlimit=per25-(1.5*iqr)
    A=df[(df[c]>lowerlimit) & (df[c]<upperlimit)]
    plt.figure()
    sns.boxplot(y=c,data=A)
    
    
```

```python
#Use standardization and perform Linear Regression
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_train=sc.fit_transform(x_train)
```

```python
x_train
```

```python
x_test=sc.transform(x_test)
```

```python
#Perform Linear Regression 
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
reg.intercept_
```

```python
y_predict=reg.predict(x_test)
```

```python
#Evaluate the model using MSE,RMSE and R Sqared
```

```python
from sklearn import metrics
```

```python
MSE=metrics.mean_squared_error(y_test,y_predict)
```

```python
RMSE=np.sqrt(MSE)
```
